```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datatsets/dailyActivity_merged.csv', parse_dates=['ActivityDate'])
```

```python
df
```

```python
to_drop = ['LoggedActivitiesDistance']
df.drop(to_drop, inplace=True, axis=1)
```

```python
df.groupby('Id').ActivityDate.min()
```

```python
df1 = pd.DataFrame(index=pd.date_range('2022-10-05', periods=50)).assign(value=1)
df2 = pd.DataFrame(index=pd.date_range('2022-10-09', periods=50)).assign(value=1)
df3 = pd.concat([df1,df2], keys=['a','b'])
df3 = df3.rename_axis(['user','date']).reset_index()
df3.groupby('user').apply(lambda x : x.resample('7D', on='date').sum())
```
