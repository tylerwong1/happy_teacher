```python
df = df.loc[~((df['AvgBreathsPerMinute'] == 0) | (df['LightSleepAvgBreathsPerMinute'] == -1.0) | (df['DeepSleepAvgBreathsPerMinute'] == -1.0) | (df['REMSleepAvgBreathsPerMinute'] == -1.0))]
df
```

```python
df['SleepDay'] = pd.to_datetime(df['SleepDay'])
```

```python
df
```

```python
weekly = df.groupby(by=['Id', pd.Grouper(key='SleepDay', freq='W')]).mean()
```

```python
weekly
```

```python
bWeek = df[df.SleepDay.dt.weekday < 5]
bWeek
```

```python
convert = bWeek.groupby(by=['Id', pd.Grouper(key='SleepDay', freq='W')]).mean().reset_index()
```

```python
convert
```

```python
start = convert.iloc[[0,1,4,15,17,26,36,42,53,61,73,76,82,91,96,98,106,109,116,123]]

num_days_intervention = [10,18,82,35,65,88,64,87,81,81,52,80,65,63,15,62,22,58,69,44]
days_after_consent = [10,6,7,18,3,24,18,4,25,1,31,15,2,3,3,13,3,8,11,14]

start.insert(2, '#DaysofIntervention', num_days_intervention)
start.insert(3, 'DaysAfterConsent', days_after_consent)
start
```

```python
end = convert.iloc[[0,3,14,16,25,35,41,52,60,72,75,81,90,95,97,105,108,115,122,126]]
end.insert(2, '#DaysofIntervention', num_days_intervention)
end.insert(3, 'DaysAfterConsent', days_after_consent)
end
```

```python
with pd.ExcelWriter('Breathing Rate Data Analysis.xlsx') as writer:
    start.to_excel(writer, sheet_name='First Week')
    end.to_excel(writer, sheet_name='Last Week')
```
