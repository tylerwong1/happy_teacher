```python
import pandas as pd
import numpy as np
pd.options.display.max_rows = 100
```

```python
# dataframe where data from fitabase is called into
df = pd.read_csv('pilot_datatsets/dailyActivity_merged.csv')
```

```python
# dataframe with the consent dates of each participant
df2 = pd.read_excel('Fitabase and Consent Dates File 2.1.23.xlsx')
```

```python
df2_sorted = df2.sort_values('ID')
```

```python
# pulls specific column desired
unique_ids = df['Id'].unique()
dfs = {}
dfs_consent = {}
for id in unique_ids:
    curr_df = df.query("Id == @id")[["ActivityDate", "TotalDistance"]]
    curr_df['ActivityDate'] = pd.to_datetime(curr_df['ActivityDate'])
    consent_df = df2_sorted.query("ID == @id")[["Consent Date"]]
    consent_df['Consent Date'] = pd.to_datetime(consent_df['Consent Date'])
    dfs[id] = curr_df
    dfs[id].reset_index(drop=True, inplace=True)
    dfs_consent[id] = consent_df
```

```python
# calculates the number of days the first data point is after the consent date
for id in unique_ids:
    consent_date = dfs_consent[id]["Consent Date"].iloc[0]
    for  i in range(dfs[id]["ActivityDate"].size):
        activity_date = dfs[id]["ActivityDate"].iloc[i]
        dfs[id]["ActivityDate"].iloc[i] = activity_date - consent_date
        days_int = int(dfs[id]["ActivityDate"][i].days)
        dfs[id]["ActivityDate"].iloc[i] = days_int
        
```

```python
# makes all dfs have 100 rows
for id in unique_ids:
    empty_rows = pd.DataFrame(index=range(100 - dfs[id].shape[0]))
    dfs[id] = pd.concat([dfs[id], empty_rows], ignore_index=True)
    dfs[id] = dfs[id].shift(dfs[id]["ActivityDate"][0])
    dfs[id].fillna(0, inplace=True)
```

```python
# retrieves only the column with the data desired
for id in unique_ids:
    curr = dfs[id]["TotalDistance"]
    dfs[id] = curr
# combines all the dfs in the dictionary into one df
combined_df = pd.concat(dfs, axis=1)
combined_df
```

```python
# replaces any NaN values with 0
combined_df.fillna(0, inplace=True)
```

```python
# name of new file that would be created/appended
file_name = 'test.xlsx'
```

```python
with pd.ExcelWriter(file_name, mode='a') as writer:
    combined_df.to_excel(writer, sheet_name='test2', index=True)
```
