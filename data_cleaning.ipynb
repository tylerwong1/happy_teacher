```python
df2_sorted = df2.sort_values('ID')
```

```python
unique_ids = df['Id'].unique()
dfs = {}
dfs_consent = {}
for id in unique_ids:
    curr_df = df.query("Id == @id")[["ActivityDate","TotalSteps"]]
    curr_df['ActivityDate'] = pd.to_datetime(curr_df['ActivityDate'])
    consent_df = df2_sorted.query("ID == @id")[["Consent Date"]]
    consent_df['Consent Date'] = pd.to_datetime(consent_df['Consent Date'])
    dfs[id] = curr_df
    dfs[id].reset_index(drop=True, inplace=True)
    dfs_consent[id] = consent_df
#unique_ids
```

```python
#dfs[10103]
#dfs_consent[10103]["Consent Date"]
#dfs[10102]["ActivityDate"]
#dfs_consent[10101]["Consent Date"]
#dfs[10102]["ActivityDate"][0] = dfs[10102]["ActivityDate"].iloc[0] - dfs_consent[10102]["Consent Date"].iloc[0]
```

```python
for id in unique_ids:
    consent_date = dfs_consent[id]["Consent Date"].iloc[0]
    for  i in range(dfs[id]["ActivityDate"].size):
        activity_date = dfs[id]["ActivityDate"].iloc[i]
        dfs[id]["ActivityDate"].iloc[i] = activity_date - consent_date
        days_int = int(dfs[id]["ActivityDate"][i].days)
        dfs[id]["ActivityDate"].iloc[i] = days_int
        
```

```python
for id in unique_ids:
    empty_rows = pd.DataFrame(index=range(100 - dfs[id].shape[0]))
    dfs[id] = pd.concat([dfs[id], empty_rows], ignore_index=True)
    dfs[id] = dfs[id].shift(dfs[id]["ActivityDate"][0])

dfs[10101]
```
